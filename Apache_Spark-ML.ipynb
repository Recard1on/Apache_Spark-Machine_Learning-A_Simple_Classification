# null kontrolü yapalım

```python
for degisken in df3.columns:
  bos = df3.filter(df3[degisken].isNull()).count()
  print(degisken + " değişkeninde " + str(bos) + " adet boş veri var.")
```

# ? işaretleri ile ilgili problemlerimiz vardı şimdi onlarla ilgilenelim.

```python
for degisken in df3.columns:
  deger = df3.filter(df3[degisken].contains("?")).count()
  print(degisken + " değişkeninde " + str(deger) + " adet ?'li veri var.")
```

###  ? olan değişkenlerin output değişkeniyle olan ilişkisine bir bakalım

```python
df3.select("workclass","occupation","native_country") \
.filter(
df3["workclass"].contains("?") |
df3["occupation"].contains("?") |
df3["native_country"].contains("?")
) \
.groupBy("workclass","occupation","native_country").count()\
.orderBy(col("count").desc())\
.toPandas().head(35)
```

### Soru işaretlerinin dağılımına baktığımızda hedef değişkenle arasında tesadüfi bir ilişki bulunuyor silmek mantıklı olacağını düşündüm.

```python
df4 = df3\
.filter(~(
    df3["workclass"].contains("?") |
    df3["occupation"].contains("?") |
    df3["native_country"].contains("?")
))
print(df3.count())
print(df4.count())
```

### Zayıf Sınıfların Kaldırılması workclass niteliğinde never-worked ve without-pay sınıfları ve occupation niteliğinde Armed-Forces sınıfı çok az tekrarlanmış. Veri setinden çıkartabiliriz.

```python
df5 = df4 \
.filter(~(df4["workclass"].contains("never-worked") |
        df4["workclass"].contains("without-pay") |
        df4["occupation"].contains("Armed-Forces") |
        df4["native_country"].contains("Holand-Netherlands") 
         ))
print(df4.count())
print(df5.count())
```

### Eğitim durumlarıyla ilgili bir düzenleme yapalım.

######         1st-4th , 5th-6th , 7th-8th : elementary-school
######         9th,10th,11th,12th : high school
######          Masters,Doctorate:Postgraduate
######          Bachelors, Some-college : undergraduate
######          bu şekilde bir birleştirme yapalım.

```python
df5 = df5 \
.withColumn("workclass", trim(col("workclass"))) \
.withColumn("education", trim(col("education"))) \
.withColumn("marital_status", trim(col("marital_status"))) \
.withColumn("occupation", trim(col("occupation"))) \
.withColumn("relationship", trim(col("relationship"))) \
.withColumn("race", trim(col("race"))) \
.withColumn("sex", trim(col("sex"))) \
.withColumn("native_country", trim(col("native_country"))) \
.withColumn("output", trim(col("output")))
```

```python
df6 = df5.withColumn("education_merged", \
      when(col("education").isin("1st-4th","education","5th-6th","education","7th-8th"), "Elementary-School")
        .when(col("education").isin("9th","10th","11th","12th"), "High-School")
        .when(col("education").isin("Masters","Doctorate"), "Post-Graduate")
        .when(col("education").isin("Bachelors","Some-college"), "Under-Graduate")
        .otherwise(col("education")))
df6.select("education","education_merged").toPandas().head(20)
```

```python
num_cols = ["age","fnlwgt","education_num","capital_gain","capital_loss","hours_per_week"]
label = ["output"]
cat_cols = ["workclass","education_merged","marital_status","occupation","relationship","race","sex","native_country"]
```

##### Gerekli kütüphaneleri ekledik 

```python
from pyspark.ml.feature import StringIndexer, OneHotEncoderEstimator, VectorAssembler
from pyspark.ml import Pipeline, PipelineModel
```

###### spark sürümünden kaynaklı olarak aşağıdaki gibi bir kod yazdım. Yeni sürümlerde böyle değil çok daha basit şekilde tek satırda halledebiliyoruz. Biraz araştırdım ama bu sürümde yapamadım.

```python
workclass_indexer = StringIndexer() \
.setInputCol("workclass") \
.setOutputCol("workclass_index") \
.setHandleInvalid("skip")
education_merged_indexer = StringIndexer() \
.setInputCol("education_merged") \
.setOutputCol("education_merged_index") \
.setHandleInvalid("skip")
marital_status_indexer = StringIndexer() \
.setInputCol("marital_status") \
.setOutputCol("marital_status_index") \
.setHandleInvalid("skip")
occupation_indexer = StringIndexer() \
.setInputCol("occupation") \
.setOutputCol("occupation_index") \
.setHandleInvalid("skip")
relationship_indexer = StringIndexer() \
.setInputCol("relationship") \
.setOutputCol("relationship_index") \
.setHandleInvalid("skip")
race_indexer = StringIndexer() \
.setInputCol("race") \
.setOutputCol("race_index") \
.setHandleInvalid("skip")
native_country_indexer = StringIndexer() \
.setInputCol("native_country") \
.setOutputCol("native_country_index") \
.setHandleInvalid("skip")

encoder = OneHotEncoderEstimator() \
.setInputCols(["workclass_index","education_merged_index","marital_status_index","occupation_index","relationship_index",
               "race_index", "native_country_index"]) \
.setOutputCols(["workclass_index_enc","education_merged_index_enc","marital_status_index_enc","occupation_index_enc",
                "relationship_index_enc", "race_index_enc", "native_country_index_enc"])
```

###### VectorAssembler, label_indexer, StandardScale işlemlerini yaptık.

```python
vector_assembler = VectorAssembler().setInputCols(num_cols + encoder.getOutputCols()).setOutputCol("vectorized_features")
```

```python
label_indexer = StringIndexer() \
.setInputCol("output") \
.setOutputCol("label")
```

```python
from pyspark.ml.feature import StandardScaler 
scaler = StandardScaler(inputCol="vectorized_features",outputCol="features", withStd=True, withMean=True)
```

###### basit bir uygulama olduğu için logisticregression algoritmasını kullandım.

```python
from pyspark.ml.classification import LogisticRegression
```

```python
lr_object = LogisticRegression() \
.setFeaturesCol("features") \
.setLabelCol("label") \
.setPredictionCol("prediction")
```

```python
train_df, test_df = df8.randomSplit([0.8, 0.2], seed=142)
```

###### pipeline oluşturup modelimi eğitip test ettim.

```python
pipeline_nesnesi = Pipeline() \
.setStages([workclass_indexer, education_merged_indexer, marital_status_indexer, occupation_indexer, relationship_indexer, 
            race_indexer, native_country_indexer, encoder, vector_assembler, label_indexer, scaler, lr_object])
```

```python
pipeline_model = pipeline_nesnesi.fit(train_df)
```

```python
sonuc_df = pipeline_model.transform(test_df).select("label","prediction")
sonuc_df.toPandas().head()
```

```python
from sklearn.metrics import accuracy_score
y_pred = sonuc_df.select("prediction").toPandas()
y_true = sonuc_df.select("label").toPandas()
```

```python
accuracy_score(y_true, y_pred)
```
